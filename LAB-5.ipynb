```python
import pandas as pd

columns = [
    "Sex","Length","Diameter","Height",
    "Whole_weight","Shucked_weight",
    "Viscera_weight","Shell_weight","Rings"
]
df = pd.read_csv("abalone.data", names=columns)
```

```python
print("Number of rows:", len(df))
print("\nColumn names:")
print(df.columns)
print("\nFirst 5 rows:")
print(df.head())
```

```python
y = df["Rings"] + 1.5
```

```python
y = (df["Rings"] + 1.5).values
X = df[["Length","Diameter","Shell_weight"]].values
```

```python
import numpy as np
np.random.seed(42)

indices = np.random.permutation(len(X))
split = int(0.8 * len(X))
train_idx = indices[:split]
test_idx = indices[split:]
X_train, X_test = X[train_idx], X[test_idx]
y_train, y_test = y[train_idx], y[test_idx]

print("Train size:", len(X_train))
print("Test size:", len(X_test))
```

```python
mean = X_train.mean(axis=0)
std = X_train.std(axis=0)
X_train = (X_train - mean) / std
X_test = (X_test - mean) / std
```

```python
def predict(X, w, b):
    return X.dot(w) + b
```

```python
def mean_squared_error(y, y_hat):
    return ((y_hat - y) ** 2).mean()
```

```python
def compute_gradients(X, y, y_hat):
    N = len(y)
    dw = (2/N) * X.T.dot(y_hat - y)
    db = (2/N) * np.sum(y_hat - y)
    return dw, db
```

```python
def update_parameters(w, b, dw, db, lr):
    w = w - lr * dw
    b = b - lr * db
    return w, b
```

```python
w = np.zeros(X_train.shape[1])
b = 0
lr = 0.01
epochs = 1000
for epoch in range(epochs):
    y_hat = predict(X_train, w, b)
    loss = mean_squared_error(y_train, y_hat)
    dw, db = compute_gradients(X_train, y_train, y_hat)
    w, b = update_parameters(w, b, dw, db, lr)

    if epoch % 100 == 0:
        print(f"Epoch {epoch}, Loss: {loss:.4f}")
```

```python
y_pred = predict(X_test, w, b)
test_loss = mean_squared_error(y_test, y_pred)
print("Test MSE:", test_loss)
for i in range(5):
    print("Predicted:", round(y_pred[i],2), " Actual:", y_test[i])
```
